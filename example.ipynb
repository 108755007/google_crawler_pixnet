```python
from pixnet import pixnet
import pandas as pd
import jieba
from gensim.models import word2vec
from tqdm import tqdm
```

```python
keyword_list=['蛋白質','雞胸肉','鱈魚']
```

main(keyword_list,num=,pages=)
num=一頁的數量
pages=總共的頁數
總共的數量=num*pages
預設num=10,pages=1

```python
data=pixnet().main(keyword_list,pages=5)
```

```python
data['蛋白質']['article']
```

```python
data['蛋白質']['keywords']
```

```python
seed = 666
sg = 1
window_size = 20
vector_size = 250
iter_=10
min_count = 3
workers = 10
epochs = 100
batch_words = 100
for keyword in tqdm(keyword_list):
    user_dict = pd.read_csv('user_dict.txt').iloc[:, 0].tolist()
    with open('user_dict.txt', 'a+', encoding='utf-8') as f:
        if keyword not in user_dict:
            f.write(f'\n{keyword}')
        for key_w in data[keyword]['keywords']:
            if key_w not in user_dict:
                f.write(f'\n{key_w}')
        f.close()
    words = [line.rstrip('\n') for line in open('add_words.txt', 'r', encoding='utf-8').readlines()]
    for word in words:
        jieba.add_word(word)
        jieba.suggest_freq(word, tune=True)
    stopwords = pd.read_csv('stop_words.txt').iloc[:, 0].tolist()
    jieba.load_userdict('user_dict.txt')
    act_ = ' '.join(data[keyword]['article']).replace('，', ' ').replace('。', ' ').replace('!', ' ').replace('?', ' ').replace('...', ' ')
    act = [i for i in jieba.cut(act_) if i not in stopwords]
    act = [word for word in act if word != ' ' and not str.isdigit(word)]
    train_data = ' '.join(act)
    with open('train_data.txt', 'w', encoding='utf-8') as new_f:
        new_f.write(train_data)
    train_data = word2vec.LineSentence('train_data.txt')
    model = word2vec.Word2Vec(
        train_data,
        min_count=min_count,
        vector_size=vector_size,
        workers=workers,
        epochs=epochs,
        window=window_size,
        sg=sg,
        seed=seed,
        batch_words=batch_words)
    try:
        pred=model.wv.most_similar(keyword, topn=30)
        print('模型預測的關鍵字:\n')
        for i,j in pred:
            if len(i)>=2:
                print(i,j)
        print('痞客幫的關鍵字:\n')
        print(data[keyword]['keywords'])
    except:
        print('爬蟲回來關鍵字小於3')
        keyword=input('請提供相關或者更短的關鍵字預測\n')
        try:
            pred=model.wv.most_similar(keyword, topn=30)
            print('模型預測的關鍵字:\n')
            for i,j in pred:
                if len(i)>=2:
                    print(i,j)
            print('痞客幫的關鍵字:\n')
            print(data[keyword]['keywords'])
        except :
            print (f"'{keyword}'此關鍵字也少於3,可能是痞客幫文章過少,換個關鍵字搜尋")
```
